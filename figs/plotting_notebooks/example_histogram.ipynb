# Exmmple histogram
Thi
s notebook is for generating an illutrative example histogram of pairwise pharmacophore distributions.

```python
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
```

```python
#peak 1
mu = 2
sigma = 1.5
peak1 = np.random.normal(mu, sigma, 200)

# peak 2
mu = 8
sigma = 1.0
peak2 = np.random.normal(mu, sigma,200)

# peak 3
mu = 12  
sigma = 0.5
peak3 = np.random.normal(mu, sigma, 100)
dist = np.hstack([peak1, peak2, peak3])


kde = stats.gaussian_kde(dist)
xx = np.linspace(0, 15, 1000)
```

```python
import seaborn as sns

sns.set(font_scale=1.4)
sns.set_style('ticks')
sns.set_palette(sns.color_palette('Paired'))
fig, ax = plt.subplots(figsize=(8, 6))

count, bins, ignored = ax.hist(dist, 30, density=True, label='Observed')
ax.plot(xx, kde(xx), label='Fit')
plt.xlim(left=0)
plt.legend()
plt.title('Aromatic-Aromatic Pairwise Distance Distribution')
plt.xlabel(r'Distance ($\AA$)')
plt.ylabel('Probability Density')
ax.set(yticklabels=[])
sns.despine()
```
